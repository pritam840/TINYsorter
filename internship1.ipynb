```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
df=pd.read_csv("/content/Iris[1].csv")
df.head()
```

```python
df.info()
```

```python
le=LabelEncoder()
df['Species'] = le.fit_transform(df['Species'])
df.head()
```

```python
x=df.drop('Species',axis=1)
y=df['Species']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
Accuracy=(y_test, y_pred)
```

```python
Accuracy
```

```python
classification_report=(y_test, y_pred)
classification_report
```

```python
confusion_matrix=(y_test, y_pred)
confusion_matrix
```

```python
confusion_matrix=(y_test, y_pred)
class_names = np.unique(y_test)
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Reds',
            xticklabels=class_names, yticklabels=class_names)
```

```python
plt.xlabel('Predicted')
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=200),
    "SVM (Linear)": SVC(kernel='linear'),
    "SVM (RBF)": SVC(kernel='rbf'),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Random Forest": RandomForestClassifier()
}
```

```python
for name, model in models.items():
    model.fit(x_train, y_train)
    score = model.score(x_test, y_test)
```

```python
print(f"{name}: Test Accuracy = {score:.4f}")
```

```python
models = {
    'Logistic Regression': LogisticRegression(max_iter=200),
    'KNN': KNeighborsClassifier(),
    'SVM': SVC(),
    'Random Forest': RandomForestClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Naive Bayes': GaussianNB()
}
```

```python
result=[]
```

```python
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    result.append({'Model': name, 'Accuracy': acc})
```

```python
results_df = pd.DataFrame(result).sort_values(by='Accuracy', ascending=False)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=results_df, x='Model', y='Accuracy', palette='Blues_d')
plt.title("Model Accuracy Comparison on Iris Dataset")
plt.ylim(0.8, 1.0)
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--')
plt.show()
```
